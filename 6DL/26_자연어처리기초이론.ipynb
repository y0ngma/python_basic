# 딥러닝 발전 분야

## 이미지 분류
  - 딥러닝으로 가장 빨리 사용되어서 성과를 낸 분야
  - 이미지 인식대회, ResNet은 150층까지 설계해서 구현
  - 대회가 없는 이유 -> 사람의 인지 능력을 추월햇지만, 한계도 있어서 -> 아직도어려운 분야
  - 트랜드가 이미지 분류 -> 이미지 생성으로 옮겨갔다
  - CNN을 기반으로 작업 => 생성을 위해서 GAN(적대적 신경망)으로 많이 사용된다.

## 음성 인식
  - 딥러닝을 적용하여 상용화까지 성공한 분야
  - 과거
    - GMM(음소인식) + HMM(시퀀셜정보알고리즘) + AM(HMM을 통과한 데이터를 학습해서 만든 음향모델) + LM(n-gram을 기반한 언어모델) + WFST(최종결합)를 결합해서 구현, 복잡, 시스템 설계도 복잡 => ASR(자동음성인식)
  - 현재
    - DNN(딥러닝대체) -> 음향모델은 시계열 데이터를 처리하는 LSTM으로 대체 + end-to-end(딥러닝의 특징) 방식 적용

## 기계 번역
  - 자연어 처리의 최종단계, 궁극의 결정체
  - 딥러닝으로 인해 정복되었다(단, 제품별로 성능이 상이하다)
  - history
    - 1950s : RBMT(규칙기반기계번역)
    - 1990s : SMT(통계기반 기계번역(단어기반,구문기반)
    - 2015~ : 신경망

## 생성 모델 학습 (최근 2-3년내에 가장 유행하는 분야, 최신 트렌드)
  - 이미지 분류, 텍스트분류, 판독하고 => 판별 모델 학습
  - 현재는 생성모델 학습에 집중, 
    - 데이터가 있다 
    - X에 대한 Y가 남자, 여자, 누구, 기타... 찾는것이 관심사가 아니라, X의 분포에 집중에서 X자체를 묘사하는 방향으로 진행 -> 적대적학습(GAN), 변분오토인코더(VAE)등이 발전하고 있다

# 자연어 처리

## 과거 
- 복잡, 무겁다, 구현도 어렵다. 시스템구성 복잡
- 오차 전파 현상 : 각 단계에서 발생되는 오차가 중첩, 가중되서 뒤로 전파되는 현상

## 딥러닝
- 자연어 처리의 최신 주류기법으로 사용
- 단어의 임베딩(백터화, 0,1로표현하는, 값으로 표현 1,2,3,,,)을 통해서 연속적인 백터로 표현이 가능 (단어를 수치로 표현할수 있게 되었다)
- 모호성, 유의성 해결 가능
- end-to-end 모델을 사용함으로 성능을 향상
- 딥러닝 도입 이후 RNN이 주류였으나, 단점을 보완한, LSTM, GRU등 활용이 고도화되었다 -> 어텐션의 등장 긴 길이의 시퀀스 처리도 쉽게 훈련이 가능해졌다

## 모호성
- 축구공을 **차**던 **차**에 **차**가 마시고 싶어서 카페에 갈려고 **차**안에서 그에게 고백하다 **차**였다
  - 파파고 : I was kicking a soccer ball and wanted to drink a cup of tea, so I tried to go to a cafe and confessed to him in the car.
  - 구글 : I confessed to him in the car to go to the cafe because I wanted to drink it.
  - 카카오 : I got kicked in the car, confessing to him in the car to go to a cafe because I wanted to drink a car from a soccer ball
  - 빙(MS) : He confessed to him in the car that he wanted to drink in the car he was playing football with, so he tried to go to the café.
- 문장내의 정보 부족으로 모호성이 발생한다. 
- 동음 이의어에 대한 문제(약한 문제)

## CNN
- 2014, 윤김
- 고정관념을 깼다( 페러다임의변화를 유도)
- 고정관념
  - 이미지는 CNN, 텍스트는 RNN이라는 고정적인 방법론을 깨고, 
  - CNN을 텍스트 분류모형에 적용하여 더 좋은 성능 냈다.
  - 형태소 분석, 문장 파싱, 개체명인식, 등등 해결책으로 제시
  - 텍스트cnn.png

## seq2seq, attention
- 2014
  - seq2seq 기법 개발
  - attention 기법 개발
  => 자연어 생성 기법 NLG(Natural Language Generation) =< 기계번역, 요약, 챗봇등이 등장

## memory
- 연속적인 방법으로 정보를 읽고, 쓰는 기법에 촛점이 맞춰져 있다
- 메모리 증강 신경망(MANN) => QA에 적합
  - 원하는 정보를(궁금한 사항들을) 신경망을 통해서 저장하고, 필요할때 조합해서 생성하여 제공한다
