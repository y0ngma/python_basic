```python
import numpy as np
import pandas as pd
import seaborn as sns
COLORS = sns.color_palette()
%matplotlib inline
```

```python
dtypes = {
    'UniPrice':np.float32,
    'CustomerID':np.int32,
    'Quantity':np.int32
}
retail = pd.read_csv('./Data/onlineRetailCleansed.csv')
retail.info()
retail.tail(2)
```

```python
retail['InvoiceDate'] = pd.to_datetime( retail['InvoiceDate'],
                                       infer_datetime_format=True )
retail.InvoiceDate.dtype
```

```python
# 전체 매출
total_revenue = retail['CheckoutPrice'].sum()
total_revenue
```

```python
# 국가별 매출
# rev_by_countries = retail.groupby('Country').sum()
rev_by_countries = retail.groupby('Country').sum()['CheckoutPrice'].sort_values()
rev_by_countries.tail()
```

```python
plot = rev_by_countries.plot( kind='bar', color=COLORS[-1], figsize=(20,10) )
plot.set_xlabel('Country', fontsize=11)
plot.set_ylabel('Revenue', fontsize=11)
plot.set_title( 'Revenue by Country', fontsize=13 )
plot.set_xticklabels( labels=rev_by_countries.index, rotation=45 )
```

```python
# 함수화 시키기 위해 받는인자로 내용 바꿔주기
def plot_bar(df, xlabel, ylabel, title, color=COLORS[0], figsize=(20,10),rotation=45):
    plot = df.plot( kind='bar', color=color, figsize=figsize )
    plot.set_xlabel( xlabel, fontsize=11)
    plot.set_ylabel( ylabel, fontsize=11)
    plot.set_title( title, fontsize=13 )
    plot.set_xticklabels( labels=df.index, rotation=rotation )
```

```python
# 월별 매출
retail.InvoiceDate.sort_values(ascending=False)
```

```python
def extract_month(date):
    month = str(date.month)
    if date.month < 10:
        month = '0' + month # yyyy-01(월) 으로 변경됨
    return str( date.year ) + month
```

```python
rev_by_month = retail.set_index('InvoiceDate').groupby(extract_month).sum()
rev_by_month.head(2)
```

```python
# 월별매출
# plot_bar(rev_by_month, 'Month', 'Revenue', 'Revenue by Month')
```

```python
# 요일별 매출
rev_by_dow = retail.set_index('InvoiceDate').groupby(
                lambda date: date.dayofweek).sum()['CheckoutPrice'].sort_values()
rev_by_dow
```

```python
DAY_OF_WEEK = np.array(['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun'])
rev_by_dow.index = DAY_OF_WEEK[rev_by_dow.index]
```

```python
rev_by_dow
```

```python
plot_bar(rev_by_dow, 'DOW', 'Revenue', 'Revenue by DOW')
```

```python
# 시간별 매출
rev_by_hour = retail.set_index('InvoiceDate').groupby(lambda date:date.hour
                         ).sum()['CheckoutPrice']
plot_bar(rev_by_hour, 'Hour', 'Revenue', 'Revenue by Hour')
```

### 상위 10 판매 품목

```python
# Top 10 판매제품
top_selling = retail.groupby(
    'StockCode').sum()['Quantity'].sort_values(ascending=False)[:3]
top_revenue = retail.groupby(
    'StockCode').sum()['CheckoutPrice'].sort_values(ascending=False)[:10]
```

```python
# top 3 아이템의 월별 판매량 추이
monthly_top3 = retail.set_index('InvoiceDate').groupby(['StockCode'
        , extract_month]).sum()[['Quantity', 'CheckoutPrice']].loc[top_selling.index]
plot_bar(monthly_top3['CheckoutPrice'], 
         'Product/Month', 'Revenue', 'Revenue of top 3 items')
```
