## 파이선 기본 문법 목록

- code -> markdown 은 주석? 이다. #의 개수로 글짜크기 설정
- 타입
    - 단일 데이터 타입 (값이 1개)
        - 수치형 : 
            - 정수 
                - 십진수, 이진수(블린), 8진수, 16진수(0xFF) 
            - 부동소수 (GPS 위도경도, -3.14, 0.0, 1.2)
        - 문자열(단일형으로 분류하지만, 실체는 연속형)
            - """문자열"""
            - "helloworld"(둘다 맞음: 파이선)
            - 'helloworld'(하나만 지원 : 자바, C)
        - 블린 boolean
            - True(1), False(0)
    - 연속 데이터 타입 (여러개의 값)
        - 리스트    :[], 순서(인덱스)가 있다. 중복값허용
        - 딕셔너리  :{}, 순서없음. 키와 값의 세트로 구성, 키중복x
        - 튜플      :(), 순서(인덱스)가 있다. 중복값허용, 단순묶음, 수정불가(읽기용)
        
        - 집합      :{}, 중복불가(불가, 제거)(~12월 26일)
        
        === 데이터 분석 (2차원)
        - 배열
        - 시리즈
        - 데이터프레임 
        
        === 딥러닝
        - 텐서 (N차원. 행렬연산)
- 조건문, 반복문 제어문 식
    - 조건식 결론 => 참이냐 거짓이냐?
        - False 상황: 0, 0.0, [], {}. (), None, 기타 조건문의 결과(비교 등)
    - 조건식을 가지고 상황따라 움직이는 방식 : 조건문
        - if
    - 반복문 : 반복작업
        - for, while
